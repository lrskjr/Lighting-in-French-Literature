```python
# The function called 'get_sentenses' takes a DataFrame and a search term as input.
# It runs through each novel and returns the sentences that holds the search term 

def find_word(search_term, text):
    # Create a regular expression pattern to match the word with optional plural forms
    pattern = r'\b{}\b'.format(re.escape(search_term))

    # Search for the pattern in the text
    if re.search(pattern, text, re.IGNORECASE):
        return True
    else:
        return False 
    
    
def get_sentenses(data_frame, search_term):
         
    # Iterate over rows using iterrows()
    for index, row in data_frame.iterrows():
        sent_list = row['text'].split('.')
        for sent in sent_list:
            search = find_word(search_term, sent)
            # Check if the search term exists in the current paragraph
            if search == True:
                # Retrieve author and title information from the current row
                au = row['author']
                ti = row['title']
                # Print a separator line
                print('*' * 20)
                # Print a message indicating where the search term was found
                print(f'In {ti.upper()} by {au.upper()}, the word {search_term.upper()} is in this sentence:\n\n {sent}')
                print('\n')

# Call the 'get_paragraphs' function with the DataFrame 'df' and the search term 'nightingale'
get_sentenses(df, 'battement de cœur') 
```
